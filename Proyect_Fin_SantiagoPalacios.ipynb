```python
# Heart Failure Prediction
# Como primer paso cargamos las librerias necesarias para ejecutar mi proyecto
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Importamos el clasificador del arbol de deciciones
from sklearn.model_selection import train_test_split # Importamos la funcion de Division 
from sklearn import metrics #Importe el módulo de métricas scikit -learn para el cálculo de precisión
```

```python
# Llamamos al archivo CSV  de la ruta almacenada en mi PC
pima = pd.read_csv("D:\\Users\\san-p\\Desktop\\PROY_FINAL\\heart_failure_clinical_records_dataset.csv")
```

```python
# Impromimos la informacion mediante la variable pima
pima.head()
```

```python
# Declaramos las variables en los ejes x, y 
feature_cols = [ 'age','anaemia','creatinine_phosphokinase', 'diabetes', 'ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','sex','smoking','time']
X = pima[feature_cols] # Caracteristicas
y = pima.DEATH_EVENT # Objetivo
```

```python
# Imprimimos o mostramos la informacion de la variable (y) para determinar el tipo de dato y las caracteristicas del mismo
y
```

```python
# con esto dividimos el conjunto de datos en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% 
```

```python
# Creamos el objeto clasificador Árbol de decisión
clf = DecisionTreeClassifier()

# Train Clasificador de árboles de decisión 
clf = clf.fit(X_train.values,y_train) #sc cabeceras

# Predecimos la respuesta para el conjunto de datos de prueba
y_pred = clf.predict(X_test.values)
```

```python
# Impromimos la presision 
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn import tree
tree.plot_tree(clf)
```

```python
# Creamos el objeto clasificador Árbol de decisión
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Train Clasificador de árboles de decisión 
clf = clf.fit(X_train.values,y_train) #values sc para cabecera

#Predecir la respuesta para el conjunto de datos de prueba
y_pred = clf.predict(X_test.values) 

#Exactitud del modelo, ¿con qué frecuencia es correcta el clasificador?(femenino)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
